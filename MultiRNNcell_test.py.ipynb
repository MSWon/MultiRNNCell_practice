```python
import tensorflow as tf
import numpy as np
```

```python
n_steps = 2
n_inputs = 3
n_hidden = 5
n_layers = 3
```

```python
X = tf.placeholder(tf.float32, [None, n_steps, n_inputs])

X_batch = np.array([
        # step 0      step 1
        [[0, 1, 2], [9, 8 ,7]], # 1st data
        [[3, 4, 5], [0, 0 ,0]], # 2nd data
        [[6, 7, 8], [6, 5, 4]], # 3rd data
        [[9, 0, 1], [3, 2, 1]], # 4th data
        ])
```

```python
cell = [tf.contrib.rnn.LSTMCell(num_units = n_hidden) for _ in range(n_layers)]
enc_cell = tf.nn.rnn_cell.MultiRNNCell(cell)

outputs, enc_states = tf.nn.dynamic_rnn(enc_cell, X, dtype = tf.float32)
```

```python
init = tf.global_variables_initializer()

with tf.Session() as sess:
    
    sess.run(init)
    
    p,q = sess.run([outputs, enc_states] , feed_dict = {X: X_batch})
    
    print("######### outputs ######### \n", p)
    print("\n######### enc_states ######### \n", q)
    print("\n######### enc_states[-1] ######### \n", q[-1])
    print("output shape:", np.shape(p))
    print("enc_states shape:", np.shape(q))
    print("enc_states[-1]:", np.shape(q[-1]))
```
